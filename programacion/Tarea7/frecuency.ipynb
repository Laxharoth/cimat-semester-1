```python
import pandas as p
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

Lena

```python
df = p.read_csv("len_out.csv",delimiter=',')
df["tamano_bytes"] = df["tamano esperado"]/8
total_bits = sum(df["tamano esperado"])
decompress_size = 262_160
compress_size   = 2_377_843
pixels = 512 * 512
print(f"tamaño original:{decompress_size}")
print(f"Longitud de código esperada:{total_bits}")
print(f"razon de compresion:{decompress_size/compress_size}")
print(f"bpp:{compress_size/pixels}")
```

```python
sns.set_context('paper')
f, ax = plt.subplots(figsize = (6,15))
sns.set_color_codes('muted')
sns.barplot(x = 'byte', y = 'tamano_bytes', data = df,
            label = 'tamaño (bytes)', color = 'red', edgecolor = 'w')
sns.set_color_codes('pastel')
sns.barplot(x = 'byte', y = 'frecuencia', data = df,
            label = 'frecuencia', color = 'b', edgecolor = 'w')
ax.legend(ncol = 2, loc = 'lower right')
sns.despine(left = True, bottom = True)
plt.show()
f.delaxes(ax)
```
